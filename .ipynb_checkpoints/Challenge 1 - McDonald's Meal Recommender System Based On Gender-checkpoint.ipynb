### 3. Does ordered grilled chicken instead of crispy increase a sandwich's nutritional value?

```python
nama_kolom = [ str(i) for i in data if i != 'Category']
```

```python
items = []
for i in data['Item']:
    if i not in items:
        items.append(i)

#Mengambil index crispy dan grilled        
crispy = []
crispy_index = []
grilled_index = []
grilled = [] 
for i in data['Item']:
    for j in i.split():
        if j == 'Crispy' or j == '(Crispy':
            crispy.append(i)
            crispy_index.append(items.index(i))
        elif j == 'Grilled' or j =='(Grilled':
            grilled.append(i)
            grilled_index.append(items.index(i))
```

```python
data_nutrisi = data[nama_kolom]
data_nutrisi_crispy = data_nutrisi.loc[crispy_index]
data_nutrisi_grilled = data_nutrisi.loc[grilled_index]

data_nutrisi_crispy_mean = data_nutrisi_crispy.mean()
data_nutrisi_grilled_mean = data_nutrisi_grilled.mean()
```

```python
data_nutrisi_grilled_mean
```

```python
nama_kolom
```

```python
x = np.arange(len(data_nutrisi_crispy_mean))
width = 0.35

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(x - width/2, data_nutrisi_crispy_mean,width, color = 'red', label = 'Crispy',alpha = 0.6)
ax.bar(x + width/2, data_nutrisi_grilled_mean,width, color = 'green', label = 'Grilled',alpha = 0.3)
ax.set_title('Rata-rata Nutrisi Crispy vs Grilled Chicken')
ax.set_xticks(x)
ax.set_xticklabels(nama_kolom)
ax.legend()
ax.tick_params(axis='x', labelrotation=80)
#ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
fig.savefig('Rata-rata Nutrisi.png', dpi=100 , orientation = 'landscape', bbox_inches='tight')
```

```python
data_nutrisi = data[nama_kolom]
data_nutrisi_crispy = data_normalized.loc[crispy_index]
data_nutrisi_grilled =  data_normalized.loc[grilled_index]

data_nutrisi_crispy_mean = data_nutrisi_crispy.mean()
data_nutrisi_grilled_mean = data_nutrisi_grilled.mean()
```

```python
x = np.arange(len(data_nutrisi_crispy_mean))
width = 0.35

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(x - width/2, data_nutrisi_crispy_mean,width, color = 'red', label = 'Crispy',alpha = 0.6)
ax.bar(x + width/2, data_nutrisi_grilled_mean,width, color = 'green', label = 'Grilled',alpha = 0.3)
ax.set_title('Rata-rata Nutrisi Crispy vs Grilled Chicken')
ax.set_xticks(x)
ax.set_xticklabels(nama_kolom)
ax.legend()
ax.tick_params(axis='x', labelrotation=80)
#ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
fig.savefig('Rata-rata Nutrisi.png', dpi=100 , orientation = 'landscape', bbox_inches='tight')
```

### 4. What about ordering eggwhites instead of whole eggs?

```python
egg_white = []
egg_white_in = []
egg = []
egg_in = []
c = 0
for i in data['Item']:
    if 'Egg White' in i:
        egg_white.append(i)
        egg_white_in.append(c)
    if 'Egg' in i:
        egg.append(i)
        egg_in.append(c)
    c+=1
    
dfew = data.loc[egg_white_in].mean()
dfe = data.loc[egg_in].mean()
```

### 5. What is the least number of items could you order from the menu to meet one day's nutritional requirements?

```python
kolom_nutrisi = []
for i in nama_kolom:
    if 'Daily Value' in i :
        kolom_nutrisi.append(i)
    elif 'Sugars' == i:
        kolom_nutrisi.append(i)
    elif 'Protein' == i:
        kolom_nutrisi.append(i)      
```

```python
data_nutrition_provided=data[kolom_nutrisi]
daily_nutrition_needed = np.array([[100,100,100,100,100,65,100,100,90,100,100,100]])
```

```python
total_nutrition = np.array([np.zeros(12)])
n = 1
sisa = np.array([np.zeros(12)])
maks_item = data_nutrition_provided.iloc[[0]]
maks = data_nutrition_provided.iloc[0]
while n < 260:
    A = [i for i in maks.values]
    B = [i for i in data_nutrition_provided.iloc[n]]
    if A < B:
        maks = data_nutrition_provided.iloc[n]
        maks_item = data_nutrition_provided.iloc[[n]]
    n = n + 1
```

```python
# Ver 2
import random

def generate_menu(gen):
    gender = gen
    minimal = 0
    maximal = 0
    if gender == 'men':
        minimal=2000
        maximal=3000
    else:
        minimal=1600
        maximal=2400
    
        
    i = 2
    while i > 1:
        df_breakfast = data[data['Category'] == 'Breakfast']
        df_beef = data[data['Category'] == 'Beef & Pork']
        df_chicken = data[data['Category'] == 'Chicken & Fish']
        df_salads = data[data['Category'] == 'Salads']
        df_snacks = data[data['Category'] == 'Snacks & Sides']
        df_desserts = data[data['Category'] == 'Desserts']
        df_beverages = data[data['Category'] == 'Beverages']
        df_coffe = data[data['Category'] == 'Coffee & Tea']
        df_smoothies = data[data['Category'] == 'Smoothies & Shakes']

        dft = df_breakfast.loc[random.choice(df_breakfast.index.tolist())]
        dft = pd.concat([dft,df_beef.loc[random.choice(df_beef.index.tolist())]],axis=1)
        dft = pd.concat([dft,df_chicken.loc[random.choice(df_chicken.index.tolist())]],axis=1)
        dft = pd.concat([dft,df_salads.loc[random.choice(df_salads.index.tolist())]],axis=1)
        dft = pd.concat([dft,df_snacks.loc[random.choice(df_snacks.index.tolist())]],axis=1)
        dft = pd.concat([dft,df_desserts.loc[random.choice(df_desserts.index.tolist())]],axis=1)
        dft = pd.concat([dft,df_beverages.loc[random.choice(df_beverages.index.tolist())]],axis=1)
        dft = pd.concat([dft,df_coffe.loc[random.choice(df_coffe.index.tolist())]],axis=1)
        dft = pd.concat([dft,df_smoothies.loc[random.choice(df_smoothies.index.tolist())]],axis=1)
        total = dft.T[['Calories']].sum()[0]

        if (total >= minimal) and (total <= maximal):
            i = 0
            print(dft.T[['Category','Item','Calories']])
            t_cal = dft.T['Calories'].sum()
            print(f'Gender : {gen}')
            print(f'Total Calories : {t_cal}')
        else:
            i = 2
```
